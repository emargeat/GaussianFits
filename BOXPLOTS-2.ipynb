```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('C:/Users/Margeat/Documents/Python/smFRET/data caro/datarho2017.csv',sep=';')
data.head()
#print(data)



fig, (ax1,ax2,ax3) = plt.subplots(nrows=3, ncols=1, figsize=(8,10), sharex=False)

xlabels=['Rho','RNAP', 'Rho Rif', 'RNAP Rif', 'Rho BCM', 'RNAP BCM']
sns.set_palette("colorblind")
sns.boxplot(x=data['Condition'],y=data['Aslow'], ax=ax1, order=xlabels)
sns.swarmplot(x=data['Condition'],y=data['Aslow'], ax=ax1, color='k', size=10, order=xlabels)
#ax1.set_yticks([0,1,2,3,4,5,6,7,8])
ax1.set_title('Fraction of slow species',fontsize=20)
ax1.set_ylabel('Slow fraction',fontsize=16)
ax1.set_xticklabels(xlabels, fontsize=16)
ax1.set_xlabel('',fontsize=16)

#ax2=fig.add_subplot(gs[0,1])
#fig2, ax2 = plt.subplots(figsize=(3,4.5))

ax2.set_title('Diffusion coefficient, slow',fontsize=20)
sns.boxplot(x=data['Condition'],y=data['Dslow'], ax=ax2, order=xlabels)
sns.swarmplot(x=data['Condition'],y=data['Dslow'], ax=ax2, color='k', size=10, order=xlabels)
ax2.set_ylabel('log D,slow',fontsize=16)
ax2.set_xticklabels(xlabels, fontsize=16)
ax2.set_xlabel('',fontsize=16)

#ax2.set_yticks([30,40,50,60,70,80,90])
#axs[1].set_yticks('ylimit',(0,100))
#plt.xticks(ticks=[0,1,2],labels=xlabels, fontsize=16)

ax3.set_title('Diffusion coefficient, fast',fontsize=20)
sns.boxplot(x=data['Condition'],y=data['Dfast'], ax=ax3, order=xlabels)
sns.swarmplot(x=data['Condition'],y=data['Dfast'], ax=ax3, color='k', size=10, order=xlabels)
ax3.set_ylabel('log D, fast',fontsize=16)
ax3.set_xticklabels(xlabels, fontsize=16)
ax3.set_xlabel('',fontsize=16)

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=1.1, 
                    top=0.9, 
                    wspace=0.6, 
                    hspace=0.55)
plt.show()
```
